```python
# Assuming idx is the index of the target anime
# Function to recommend similar anime
def recommend_similar_anime(target_anime, data, cosine_sim, top_n=5):
    # Get the index of the target anime
    idx = data[data['anime_title'] == target_anime].index[0]
    
    try:
        # Get similarity scores for the target anime
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Exclude the target anime from the results
        sim_scores = [(i, score) for i, score in sim_scores if i != idx]

        # Sort the scores in descending order
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Get the top_n most similar anime
        sim_scores = sim_scores[:top_n]

        # Get the anime indices
        anime_indices = [i[0] for i in sim_scores]

        # Return the top_n most similar anime titles
        return data['anime_title'].iloc[anime_indices]

    except IndexError:
        print(f"Index {idx} is out of bounds for the cosine similarity matrix.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function to recommend similar anime
recommended_anime = recommend_similar_anime('Anime A', data, cosine_sim)
print("Recommended Anime:", recommended_anime.tolist())
```
